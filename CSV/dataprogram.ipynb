```python
import os
import pandas as pd
import csv
import glob
```

```python
# CSV 파일 경로 지정(수정)
csv_file_path = "data/34/*.csv"

# CSV 파일 리스트 가져오기
csv_files = glob.glob(csv_file_path)

# CSV 파일 반복 처리
for csv_file in csv_files:

    # 새로운 CSV 파일 이름 지정(수정)
    new_csv_file = "sddata/34/new_" + os.path.basename(csv_file)

    # 새로운 폴더 생성
    new_folder = os.path.dirname(new_csv_file)
    os.makedirs(new_folder, exist_ok=True)

    # CSV 파일 열기
    with open(csv_file, encoding='utf-8', newline='') as csvfile:

        # CSV 파일 읽기
        csvreader = csv.reader(csvfile)

        # 새로운 CSV 파일 열기
        with open(new_csv_file, 'w', newline='') as new_csvfile:

            # CSV 파일 쓰기
            csvwriter = csv.writer(new_csvfile)

            # 행 수 세기
            count = 0

            # CSV 파일 반복하기
            for row in csvreader:

                # 첫 번째 행, 두 번째 행은 건너뛰기
                if count != 0 and count != 1 and count != 2 and count != 3 and count != 4 and count != 5 and count != 6:

                    # CSV 파일에 새로운 행 쓰기
                    csvwriter.writerow(row)

                # 행 수 증가
                count += 1
```

```python

# 입력 폴더 경로(수정)
input_folder = "sddata/34"

# 출력 폴더 경로(수정)
output_folder = "fixdata/34"

# 입력 폴더 내의 모든 CSV 파일을 가져옴
csv_files = [f for f in os.listdir(input_folder) if f.endswith('.csv')]

# 출력 폴더가 없으면 생성함
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# CSV 파일을 하나씩 처리함
for i, csv_file in enumerate(csv_files):
    # CSV 파일을 읽음
    df = pd.read_csv(os.path.join(input_folder, csv_file), encoding='iso-8859-1', error_bad_lines=False, header =None)
 
    # AccY 열 삭제
    df.drop([0, 1, 2, 3, 4, 5, 6, 8], axis=1, inplace=True)
    
    # 5초단위 데이터 저장
    df = df.loc[0:10001]
    
    # #행열전환(class 추가)
    df=df.transpose()
    df['10001'] = 3
    df=df.transpose()
    
    
    # 출력 파일 이름을 생성함(클레스 이름 수정)
    output_file = f"34_{i+1}.csv"
    
    # CSV 파일을 출력함
    df.to_csv(os.path.join(output_folder, output_file), index=False)
```

```python
# CSV 파일 경로 지정(수정)
csv_file_path = "fixdata/34/*.csv"

# CSV 파일 리스트 가져오기
csv_files = glob.glob(csv_file_path)

# CSV 파일 반복 처리
for csv_file in csv_files:

    # 새로운 CSV 파일 이름 지정(수정)
    new_csv_file = "lastdata/" + csv_file.split("/")[-1]

    # 디렉토리가 없으면 생성하기
    os.makedirs(os.path.dirname(new_csv_file), exist_ok=True)

    # CSV 파일 열기
    with open(csv_file, encoding='utf-8', newline='') as csvfile:

        # CSV 파일 읽기
        csvreader = csv.reader(csvfile)

        # 새로운 CSV 파일 열기
        with open(new_csv_file, 'w', newline='') as new_csvfile:

            # CSV 파일 쓰기
            csvwriter = csv.writer(new_csvfile)

            # 행 수 세기
            count = 0

            # CSV 파일 반복하기
            for row in csvreader:

                # 첫 번째 행, 두 번째 행은 건너뛰기
                if count != 0 and count != 1:

                    # CSV 파일에 새로운 행 쓰기
                    csvwriter.writerow(row)

                # 행 수 증가
                count += 1
```

```python
#데이터 프레임 만들기
df1 = pd.DataFrame() 

# 입력 폴더 경로(수정)
input_folder = "lastdata/34"

# 출력 폴더 경로(수정)
output_folder = "usedata/34"

# 입력 폴더 내의 모든 CSV 파일을 가져옴
csv_files = [f for f in os.listdir(input_folder) if f.endswith('.csv')]

# 출력 폴더가 없으면 생성함
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    # CSV 파일을 하나씩 처리함
for i, csv_file in enumerate(csv_files):
    # CSV 파일을 읽음
    df = pd.read_csv(os.path.join(input_folder, csv_file), encoding='iso-8859-1', error_bad_lines=False, header =None)
    
    #행열 전환
    df=df.transpose()
 
    # 데이터 모집
    df1 = pd.concat([df1,df])
    
# 출력 파일 이름을 생성함(수정: 클레스_데이터 횟수)
output_file = f"34_{i+1}.csv"
    
# CSV 파일을 출력함
df1.to_csv(os.path.join(output_folder, output_file), index=False)
```
